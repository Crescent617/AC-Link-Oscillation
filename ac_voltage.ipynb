```python
import numpy as np
import pandas as pd
# import seaborn as sns
import matplotlib.pyplot as plt

# ## for Palatino and other serif fonts use:
plt.style.use('seaborn-white')
```

```python
# df = pd.read_csv('./ac_voltage.csv')
df = pd.read_csv('./dps_ac.csv')
df.columns = ['t', 'a', 'b']
```

```python
fig, (p1, p2) = plt.subplots(2, 1, sharex=True, figsize=(6, 5), dpi=300)
# plt.xlabel('$t/s$')

x, u1, u2 = df['t'], df['a'], df['b']
p1.plot(x, u1, linewidth=2)
p1.grid()
p1.set_ylim(-1000, 1000)

p2.plot(x, u2, 'y', linewidth=2)
p2.grid()
p2.set_ylim(-1000, 1000)
plt.tight_layout()
```
