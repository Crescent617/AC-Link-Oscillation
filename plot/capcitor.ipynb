```python
def cut_index(df):
    pre = 0
    ans = [0]
    for i, x in enumerate(df.index):
        if x < pre:
            ans.append(i)
        pre = x
    return ans + [len(df.index)]

def my_plot(df):
    cut = cut_index(df)
    for s, e in zip(cut, cut[1:]):
        plt.plot(df.iloc[s:e, :])
```

```python
u_dc = 'V_D'

plt.figure(dpi=300, figsize=(6,3))

my_plot(c_sw[u_dc].to_frame())

# plt.legend()
# sns.despine()
plt.ylabel(r'$u_{dc2}/V$')
# plt.ylim(0, 900)
plt.xlabel(r'$t / s$')
plt.grid(True)
plt.tight_layout()
    
```
