```python
dps = dps.iloc[-3000:]
```

```python
dpi = 175

fig = plt.figure(dpi=dpi, figsize=(6,6))
# axes[0].set(title=title+'\n')

x = dps.index
u, v, i = 'Vm4', 'Vm5', 'Am3'

plt.subplot(311)
plt.plot(x, dps[u], label='$u_{ab}/V$', linewidth=1.5)
plt.ylabel('$U_{ab}/V$')#, xticks=[])
plt.xticks([])
plt.grid(True)

plt.subplot(312)
plt.plot(x, dps[v], label='$u_{cd}/V$', linewidth=1.5)
plt.ylabel('$U_{cd}/V$')#, xticks=[])
plt.xticks([])
plt.grid(True)

plt.subplot(313)
plt.plot(x, dps[i], label='$i_L/V$', linewidth=1.5)
plt.ylabel('$i_L/V$')#, xticks=[])
plt.grid(True)
plt.xlabel(f'\n$t/s$')
# plt.legend()

plt.tight_layout()
```
